```python
import pandas as pd
cleaned_file_path = '/Users/romano/Desktop/twitterBot/data/cleaned_trades_data.csv'
df = pd.read_csv(cleaned_file_path)
```

```python
import pandas as pd

# Group by 'Year_Week'
weekly_group = df.groupby('Year_Week')

# Initialize a list to store weekly performance data
weekly_performance = []

for name, group in weekly_group:
    total_trades = len(group)
    wins = group['Win'].sum()  # Sum of wins
    losses = group['Loss'].sum()  # Sum of losses
    win_percentage = (wins / total_trades) * 100 if total_trades > 0 else 0
    loss_percentage = (losses / total_trades) * 100 if total_trades > 0 else 0
    
    # Calculate the total performance using Profit(Loss) $
    total_profit_loss_dollars = group['Profit(Loss) $'].sum()

    # Calculate position size as # Shares * Price p.share
    total_position_size = (group['# Shares'] * group['Price p.share']).sum()

    # Calculate performance percentage based on the new position size
    total_performance = (total_profit_loss_dollars / total_position_size) * 100 if total_position_size > 0 else 0

    # Detailed trades breakdown
    trades_details = group.apply(
        lambda row: {
            'Ticker': row['Ticker'],
            'Buy Price': row['Price p.share'],
            'Sell Price': row['Price p.share.1'],
            'Profit(Loss) $': row['Profit(Loss) $'],
            '% Gain/Loss': (row['Profit(Loss) $'] / (row['Price p.share'] * row['# Shares'])) * 100 if row['Price p.share'] > 0 else 0
        }, axis=1
    ).tolist()

    # Append the weekly performance data
    weekly_performance.append({
        'Year_Week': name,
        'Total Trades': total_trades,
        'Win %': win_percentage,
        'Loss %': loss_percentage,
        'Total Performance (%)': total_performance,
        'Trades': trades_details
    })

# Convert to DataFrame for better display
weekly_performance_df = pd.DataFrame(weekly_performance)

# Display the DataFrame
print(weekly_performance_df)
```
